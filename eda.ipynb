# Loading The Dataset

```python
from datasets import load_dataset
import plotly.express as px

LABELS_MAPPING = {0: "method", 1: "background", 2: "result"}
```

```python
dataset = load_dataset('scicite', cache_dir='./data')
df = dataset['train'].to_pandas()
dataset['train'].features
df['label_text'] = df['label'].apply(lambda x: LABELS_MAPPING[x])
```

# Dataset Label Distribution

```python
fig = px.histogram(df, x='label_text', color='label_text')
fig.update_layout(
    title={
        'text': "SciCite Label Distribution",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)
fig.show()
```

# Context Word Count Distribution

```python
df['context_word_count'] = df['string'].apply(lambda x: len(x.split(' ')))
```

```python
fig_wc_dist = px.histogram(df, x='context_word_count')
fig_wc_dist.update_layout(
    title={
        'text': "Context Word Count Distribution",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)
fig_wc_dist.show()
```
